```python
import os
import datetime
from fillpdf import fillpdfs
import pandas as pd
```

```python
#Update these values as needed for RTK contact information and request. To change other information, update params below

name = "Your Name"
address = "Your Street Address"
city = "Your City"
state = "Your State"
zip = "Your ZIP"
email = "Your Email"
telephone = "Your Phone"

request = """(1) Copies of the most recent collective bargaining agreements between the district and all labor unions such as the teacher’s union and any support staff unions, transportation unions, or school security/police unions (if such unions exist).
\n\n(2) Copies of the most recent contract between the district and the Superintendent.
\n\n(3) Statements of Financial Interest for all School Board Directors, the Superintendent, and district’s solicitor(s) for the 2022 calendar year as specified under the Public Official and Employee Ethics Act (65 Pa.C.S. §§ 1101 - 1113)."""

output_folder = "SchoolRTK"
output_csv = "sd_contacts.csv"

if not os.path.exists(os.path.join(os.getcwd(), output_folder)):
    os.makedirs(os.path.join(os.getcwd(), output_folder))
    print("{} folder created at {}".format(output_folder, os.path.join(os.getcwd(), output_folder)))
```

```python
contacts = pd.read_csv("RTK_Contacts_Advanced.csv")
```

```python
#Note -- you could replace "School District" with any of the OOR's subtypes
#Or you could filter by county, etc...

#ALSO -- this file probably needs some manual cleanup to account for errors on OOR's website
selected_contacts = contacts[contacts["subtype"] == "School District"].copy()
```

```python
#Get today's date. If you want a different date (i.e. in the future), just write over formatted_date with a string
now = datetime.datetime.now()
formatted_date = f"{now.month}/{now.day}/{now.year}"
print(formatted_date)

for index, row in selected_contacts.iterrows():
    #print(row)
    agency = row["Agency"]

    #Filename pattern is RTK_XXXX_Agency_Name.pdf
    filename = "RTK_" + str(int(row["OOR_ID"])).zfill(4) + "_" + agency.replace(" ","_") + ".pdf"
    selected_contacts.loc[index, "filepath"] = os.path.join(os.getcwd(), "SchoolRTK", filename )
    

   #If you need to change any of the other delivery information, check the field values in the RTK.pdf file
    params = {"DateRequest" : formatted_date,
          "Name" : name,
          "StreetAddress" : address,
          "City" : city,
          "State" : state,
          "Zip" : zip,
          "Email" : email,
          "Telephone" : telephone,
          "RECORDS" : request,
          "Agency" : agency,
          "Submitted" : "Email",
          "QuestionsContact": "Email",
          "MethodContact" : "Electronic",
          "MethodCopies" : "Electronic",
          "FeeNotification" : "Other",
          "CertifiedCopies" : "No",
          "OtherFeeThreshold" : "0"
          }

    #RTK.pdf is a slightly modified version of OOR's form with simplified field names
    fillpdfs.write_fillable_pdf('RTK.pdf', os.path.join(output_folder, filename),params)

 
```
