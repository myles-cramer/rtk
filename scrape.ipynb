```python
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
rtk_contacts = pd.read_csv("RTK_Contacts_Simple.csv",
                           index_col= "OOR_ID")
```

```python
rtk_contacts.head()
```

```python
rtk_contacts.head()

for index, row in rtk_contacts.iterrows():
    url = row["Link"]
    print(row["Link"])
    
   

    page = requests.get(url=url)

    if page.status_code != 200:
        print("Page for {} returned with status code {}".format(rtk_contacts[index,"Agency"], page.status_code))
        continue

    try:
        soup = BeautifulSoup(page.content, "html.parser")

        website = soup.find("p", id="website").text
        #print(website)

        address = soup.find("p", id="address")

        for br_tag in address.find_all("br"):
            br_tag.replace_with(",")

        address = address.get_text(separator= " ")
        #print(address)

        type = soup.find("p", id="type").text
        #print(type)

        subtype = soup.find("p", id="subtype").text
        #print(subtype)

        county = soup.find("p", id="county").text
        #print(county)

        municipality = soup.find("p", id="municipality").text
        #print(municipality)

        school_district = soup.find("p", id="schoolDistrict").text
        #print(school_district)

        rtk_contacts.at[index, "website"] = website
        rtk_contacts.at[index, "address"] = address
        rtk_contacts.at[index, "type"] = type
        rtk_contacts.at[index, "subtype"] = subtype
        rtk_contacts.at[index, "county"] = county
        rtk_contacts.at[index, "municipality"] = municipality
        rtk_contacts.at[index, "school_district"] = school_district

    except Exception as e:
        print("Error with {}... {}".format(rtk_contacts[index, "Agency"], e))
        rtk_contacts.at[index, "website"] = "ERROR"
        rtk_contacts.at[index, "address"] = "ERROR"
        rtk_contacts.at[index, "type"] = "ERROR"
        rtk_contacts.at[index, "subtype"] = "ERROR"
        rtk_contacts.at[index, "county"] = "ERROR"
        rtk_contacts.at[index, "municipality"] = "ERROR"
        rtk_contacts.at[index, "school_district"] = "ERROR"

    finally:
        if index % 20 == 0:
            print("Processing OOR_ID {}".format(index))

        
        
   
```
